```python
df = pd.read_csv(r"C:\Users\ADMIN\Desktop\data analytics project\online_retail.csv", encoding= 'unicode_escape')
```

```python
df = df.astype(object)
```

```python
df = df.where(pd.notnull(df), None)
```

```python
df.isnull().sum()
```

```python
data = df.values.tolist()
```

```python
import math

cleaned_data = []

for row in data:
    cleaned_row = []
    for value in row:
        if isinstance(value, float) and math.isnan(value):
            cleaned_row.append(None)
        else:
            cleaned_row.append(value)
    cleaned_data.append(cleaned_row)
```

```python
cursor = conn.cursor()
```

```python
query = """
INSERT INTO retail_raw (
    InvoiceNo, StockCode, Description, Quantity,
    InvoiceDate, UnitPrice, CustomerID, Country
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
"""
```

```python
for i in range(0, len(cleaned_data), 500):
    batch = cleaned_data[i:i+500]
    cursor.executemany(query, batch)
    conn.commit()
    print(f"Inserted rows till: {i+len(batch)}")
```
